```python
# 1. Importar librerías
import pandas as pd
import matplotlib.pyplot as plt

# 2. Crear dataset ficticio
data = {
    'Tienda': ['Sucursal Miravalle', 'Sucursal Iztapalapa', 'Sucursal Vallejo'],
    'Unidades_Manipuladas': [1140, 1850, 1679],
    'Unidades_Dañadas': [53, 48, 67],
    'Unidades_Recuperadas': [20, 15, 25],
    'Costo_Unitario': [155, 180, 160]  # en pesos
}

df = pd.DataFrame(data)

# 3. Calcular KPIs
df['Unidades_No_Recuperables'] = df['Unidades_Dañadas'] - df['Unidades_Recuperadas']
df['Merma_%'] = (df['Unidades_No_Recuperables'] / df['Unidades_Manipuladas']) * 100
df['Tasa_Recuperacion_%'] = (df['Unidades_Recuperadas'] / df['Unidades_Dañadas']) * 100
df['Costo_Merma'] = df['Unidades_No_Recuperables'] * df['Costo_Unitario']

# 4. Mostrar resultados
print("KPIs por tienda:\n")
print(df[['Tienda', 'Merma_%', 'Tasa_Recuperacion_%', 'Costo_Merma']])

# 5. Visualización: Merma por tienda
plt.figure(figsize=(8,5))
plt.bar(df['Tienda'], df['Merma_%'], color='orange')
plt.title('Merma por Manipulación (%) por Tienda')
plt.ylabel('Porcentaje de Merma')
plt.xlabel('Tienda')
plt.show()
```
